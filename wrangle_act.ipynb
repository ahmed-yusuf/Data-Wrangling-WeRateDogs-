##### GATHERING TWEET_JSON FILE

A DataFrame is created with the three columns tweet_id, retweet_count, favorite_count.


The tweet.json file(collected from the twitter API) has been read with the open function
Values has been passed from the json file into the tweet_df and appended to the dataframe under the appropriate columns.

#### NOTE:ONLY RELEVANT COLUMNS ARE SELECTED FROM THE TWEET_JSON FILE  I.E. RETWEET_COUNTT,FAVORITE_COUNT AND TWEET_ID

```python
# Tweet IDs for which to gather additional data via Twitter's API
#Load the from the API 
#Created A new dataframe and query the tweet_json.txt file for tweet_ids,retweet_count and favorite_count
#The columns were passed into the tweet_df
tweet_df = pd.DataFrame(columns=['tweet_id', 'retweet_count', 'favorite_count'])
with open('tweet_json.txt') as f:
    for line in f:
        status  = json.loads(line)
        tweet_id = status['id_str']
        retweet_count = status['retweet_count']
        favorite_count = status['favorite_count']
        tweet_df = tweet_df.append(pd.DataFrame([[tweet_id, retweet_count, favorite_count]],
                                        columns=['tweet_id', 'retweet_count', 'favorite_count']))
tweet_df = tweet_df.reset_index(drop=True)
```
